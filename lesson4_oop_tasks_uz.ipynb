## 3-vazifa. `Counter` klassi

Hisoblagich klassini yarating:
- atribut: `value` (boshlang'ich qiymati 0)
- `increment()` metodi, 1 ga oshirish
- `reset()` metodi, 0 ga qaytarish
- `get()` metodi, joriy qiymatni qaytarish

```python
c = Counter()
c.increment()
c.increment()
c.increment()
print(c.get())   # 3
c.reset()
print(c.get())   # 0
```

## 4-vazifa. `Employee` va `Manager` ierarxiyasi

Ikkita klass yarating.

**`Employee`** (xodim):
- atributlar: `name`, `salary`
- `info()` metodi, ism va maoshni chiqaradi

**`Manager(Employee)`** (menejer, voris):
- qo'shimcha atribut: `team_size`
- `info()` ni qayta aniqlang, jamoa hajmi haqidagi ma'lumotni qo'shing
- `bonus()` metodi, maoshning 10% ini qaytaradi

```python

emp.info()
# Xodim: Ivan, maosh: 50000

mgr = Manager("Anna", 80000, team_size=5)
mgr.info()
# Xodim: Anna, maosh: 80000, jamoa: 5 kishi
```

## Uy vazifasi 1. `Book` klassi

Kitob klassini amalga oshiring:
- atributlar: `title`, `author`, `pages`, `read_pages` (boshlang'ich qiymati 0)
- `read(n)` metodi, `n` sahifa o'qish (`read_pages` `pages` dan oshmasligi uchun tekshiruv bilan)
- `progress()` metodi, o'qilgan foizni qaytaradi
- `is_finished()` metodi, agar kitob to'liq o'qilgan bo'lsa, `True` qaytaradi

```python
b = Book("1984", "Oruell", 320)
b.read(80)
print(b.progress())      # 25.0
print(b.is_finished())   # False
b.read(500)              # qoldiqdan ko'p, faqat oxirigacha o'qiymiz
print(b.is_finished())   # True
```

## Uy vazifasi 3 (yulduzcha bilan). `SimpleStatistics` klassi

Oddiy statistika klassini amalga oshiring:
- konstruktor sonlar ro'yxatini qabul qiladi
- `mean()` metodi, o'rtacha
- `median()` metodi, mediana
- `mode()` metodi, moda (eng tez-tez uchraydigan qiymat)
- `summary()` metodi, hamma uch qiymatni chiqaradi

Bu `pandas` va `numpy` dagi klasslar qanday tuzilganini tushunish yo'lidagi birinchi qadam.

```python

# mean = 3.125
# median = 3.5
# mode = 4
```
